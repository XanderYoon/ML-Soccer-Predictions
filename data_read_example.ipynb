Use SQL queries to access tables in the database and read them into pandas databases. This notebook has an example.

Import dependencies

```python
import numpy as np
import pandas as pd
import sqlite3
```

Read database

```python
path = "database.sqlite"
connection = sqlite3.connect(path)
```

```python
tables = pd.read_sql("SELECT * FROM sqlite_master WHERE type = 'table';", connection)
tables
```

```python
player = pd.read_sql("SELECT * FROM Player;", connection)
player
```
