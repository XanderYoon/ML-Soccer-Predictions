```python
team_features = ["record", "goal_diff", "oppenent_rating"]
team_attributes_features = ["buildUpPlaySpeed", "buildUpPlayPassing", "chanceCreationPassing", "chanceCreationCrossing", "defenceAggression", "defenceTeamWidth"]
player_avg_features = ["overall_rating"]
player_ind_features = ["overall_rating", "height", "weight", "sprint_speed", "stamina"]
```

```python
X_train, y_train = soccer_database.get_features_for_matches(
    train_matches, 
    team_features,
    team_attributes_features,
    player_avg_features,
    player_ind_features
)
```

```python
X_test, y_test = soccer_database.get_features_for_matches(
    test_matches, 
    team_features,
    team_attributes_features,
    player_avg_features,
    player_ind_features
)
```

```python
print(len(X_test))
print(len(X_train))

not_draw = y_train[:,1] != 1
not_draw_xtrain = X_train.iloc[not_draw]
not_draw_ytrain = y_train[not_draw,:]
not_draw_ytrain = np.argmax(not_draw_ytrain,axis=1)
not_draw_ytrain //= 2

not_draw = y_test[:,1] != 1
not_draw_xtest = X_test.iloc[not_draw]
not_draw_ytest = y_test[not_draw,:]
not_draw_ytest = np.argmax(not_draw_ytest,axis=1)
not_draw_ytest //= 2

print(not_draw_xtrain)
```

```python
param_dist = {'n_estimators': scipy.stats.randint(50,500),
              'max_depth': scipy.stats.randint(1,20)}

# Create a random forest classifier
rf = RandomForestClassifier()

# Use random search to find the best hyperparameters
rand_search = RandomizedSearchCV(rf,
                                 param_distributions = param_dist,
                                 n_iter=5,
                                 cv=5)

# Fit the random search object to the data
rand_search.fit(not_draw_xtrain.to_numpy(), not_draw_ytrain)
best_rf = rand_search.best_estimator_

# Print the best hyperparameters
print('Best hyperparameters:',  rand_search.best_params_)
```

```python
y_pred = best_rf.predict(not_draw_xtest)

# Create the confusion matrix
cm = confusion_matrix(not_draw_ytest, y_pred)

ConfusionMatrixDisplay(confusion_matrix=cm).plot();
```
